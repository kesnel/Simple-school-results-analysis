## Let's see how many student passed the examens for each Trimester

> #### Trimester 1

```python
# Make a filter to see all specified students for the first trimester
filt = (df['Trimester 1'] >= 5)
df.loc[filt, ['Nom', 'Prenom','Trimester 1']].sort_values(by='Trimester 1', ascending=False).head(3)
```
